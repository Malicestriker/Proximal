```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
input = open('a1a_Basic.txt', 'r')
b = []
for line in input:
    b.append(float(line))
b = np.array(b)
```

```python
input = open('a1a_Inertial.txt', 'r')
i = []
for line in input:
    i.append(float(line))
i = np.array(i)
```

```python
input = open('a1a_FISTA.txt', 'r')
f = []
for line in input:
    f.append(float(line))
f = np.array(f)
```

```python
input = open('a1a_Nesterov.txt', 'r')
n = []
for line in input:
    n.append(float(line))
n = np.array(n)
```

```python
maxLen = max([len(b), len(i), len(f), len(n)])
x = range(maxLen)

fig, ax = plt.subplots()
ax.plot(b, color='red', ls='--', lw=0.5, label='Basic')
ax.plot(i, color='blue', ls='--', lw=0.5, label='Inertial')
ax.plot(f[0:1200], color='gray', ls='--', lw=0.5, label='FISTA')
ax.plot(n[0:1200], color='black', ls='--', lw=0.5, label='Nesterov')
ax.legend(title="Performance on a1a")
```
